```python
# ------------------------------Gráfico de ejemplo------------------------------#
import pandas as pd

# Cargar el Excel
file_path = "C:\\Users\\Franco\\Desktop\\Grafico Excel Belen\\Grafico Belen.xlsx" # Modificar la ruta a la tuya
data = pd.read_excel(file_path)

# Mostrar las primeras 5 filas para probar
data.head()
```

```python
# ------------------------------Gráfico 1: Cantidad de repuestos por categoría------------------------------#

# Libreria que usé para crear los gráficos porque me pareció muy simple su estructura

import matplotlib.pyplot as plt

# Guarda la tabla 

repuestos_por_categoria = data['Categoría'].value_counts()

# Uso el método "figure" para configurar el gráfico
# La función "figsize" establece el tamaño de todo el gráfico. Su anchura y altura.

plt.figure(figsize=(8, 6))

# plot: crea el gráfico | Kind: Tipo de grafico (barh: gráfico de barras en horizontal) | Color: Color de las barras

repuestos_por_categoria.plot(kind='barh', color='skyblue')

# Titulo del grafico

plt.title('Cantidad de Repuestos por Categoría')

# Nombre de la etiqueta en la cordenada x. Es decir, nombre de la etiqueta que representa los datos en la cordenada x.

plt.xlabel('Categoría')

# Mismo que la línea anterior solo que influye en la cordenada y.

plt.ylabel('Cantidad de Repuestos')

# Rota las palabras/letras de los datos en la cordenada x.

plt.xticks(rotation=45)

# Grid: Configuracion del "grid". Es decir, el cuadrante se encuentran las barras
# linestyle: Estilo de línea para dividir entre los datos en la cordenada que marca la variable "axis"
# alpha: Modifica la transparencia del "linestyle" entre un rango específico (1 a 0)

plt.grid(axis='x', linestyle='--', alpha=0.5)

# Muestra el gráfico

plt.show()
```

```python
# ------------------------------Gráfico 2: Cantidad total disponible por categoría------------------------------#

cantidad_total_por_categoria = data.groupby('Categoría')['Cantidad Disponible'].sum()

# En los metodos, funciones, etc. que se repiten no puse su definicion porque es lo mismo

plt.figure(figsize=(8, 6))
cantidad_total_por_categoria.plot(kind='pie', color='lightgreen')
plt.title('Cantidad Total Disponible por Categoría')
plt.show()
```

```python
# ------------------------------Gráfico 3: Cantidad disponible de repuestos mayores a 250 divididos por categoría------------------------------#

# Filtrar los repuestos cuya cantidad disponible supera los 250

repuestos_mas_de_250 = data[data['Cantidad Disponible'] > 250]

# Agrupa los repuestos por categoría y hace la suma de la cantidad disponible de todos los repuestos separados por categoría
cantidad_disponible_mas_de_250 = repuestos_mas_de_250.groupby('Categoría')['Cantidad Disponible'].sum()

plt.figure(figsize=(8, 6))
cantidad_disponible_mas_de_250.plot(kind='bar', color='salmon')
plt.title('Cantidad Disponible de Repuestos (>250) por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad Disponible')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
